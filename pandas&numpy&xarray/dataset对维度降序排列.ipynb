## 比较 `reindex` 与 [x] 的方式的降序排列的方式


使用 `xarray.Dataset.reindex` 的方式实现降序排列,  
[xarray.DataArray.reindex](http://xarray.pydata.org/en/stable/generated/xarray.DataArray.reindex.html),  
[xarray.Dataset.reindex](http://xarray.pydata.org/en/stable/generated/xarray.Dataset.reindex.html)

### 1-1 使用 `xarray.Dataset.reindex` 的方式实现对于某个维度降序排列  
看出对某个维度降序排列，且数据也根据维度的索引变化也发生了变化，相当于只是`降序`

```python
da = xar.DataArray(
    np.arange(4),
    coords=[np.array([90, 89, 88, 87])],
    dims="lat",
)
da
```

```python
da.lat[::-1]
```

```python
da.reindex(lat=da.lat[::-1])
```

### 1-2 使用直接赋值的方式进行降序  
使用此种方式实际是`维度索引-index`保持不变，而调换了其中的`值`

```python
da['lat']=da['lat'][::-1]
da
```
